```python
train_df_new = train_df.dropna(subset='price_CHF')
train_df_new
```

```python
test_df_new = test_df
```

```python
le = LabelEncoder()
label = le.fit_transform(train_df_new['season'])
label_test = le.fit_transform(test_df_new['season'])

train_df_new['season'] = label
test_df_new['season'] = label_test
```

```python
for item in train_df_new.columns:
    train_df_new[item] = train_df_new.groupby('season')[item].transform(lambda x: x.fillna(x.median()))
for item in test_df_new.columns:
    test_df_new[item] = test_df_new.groupby('season')[item].transform(lambda x: x.fillna(x.median()))
```

```python
X_train = train_df_new.drop(['price_CHF'], axis = 1)
y_train = train_df_new['price_CHF']
X_test = test_df_new
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic, ExpSineSquared, DotProduct, ConstantKernel)
gpr = GaussianProcessRegressor(kernel=DotProduct())
gpr.fit(X_train, y_train)
```

```python
pred = gpr.predict(X_test)
```

```python
kernel = gpr.kernel_
```

```python
from sklearn.model_selection import GridSearchCV

models = GaussianProcessRegressor()
                
param_grid = [{
    "alpha":  [1e-2, 1e-3],
    "kernel": [RBF(l) for l in np.logspace(-1, 1, 2)]
}, {
    "alpha":  [1e-2, 1e-3],
    "kernel": [DotProduct(sigma_0) for sigma_0 in np.logspace(-1, 1, 2)]
}, {
    "alpha":  [1e-2, 1e-3],
    "kernel": [Matern(length_scale=1.0, nu=1.5)]
    
}]
```

```python
clf = GridSearchCV(models, param_grid)
clf.fit(X_train, y_train)
```
